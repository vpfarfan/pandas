## Carga de tabla de observaciones

```python
# Load data
obs = pd.read_csv('./observaciones_Vero.csv')
# obs['FEC_EVENT'] = pd.to_datetime(obs['FEC_EVENT'])
obs = obs.drop([obs.columns[0]], axis='columns')
obs.head()
```

```python
freq = pd.read_csv('./observaciones_Frec.csv')
freq = freq.drop([freq.columns[0]], axis='columns')

obs = pd.merge(obs, freq, how='inner')
obs.head()
```

```python
freq = pd.read_csv('./observaciones_Dev.csv')
freq = freq.drop([freq.columns[0]], axis='columns')

obs = pd.merge(obs, freq, how='inner')
obs.head()
```

## Dataset para entrenar el modelo
Entrenamos con los datos de los meses 1 a 9 de 2018, para luego predecir en el último trimestre del 2018

Features:
- promedio de sesiones por mes
- cantidad de conversiones anteriores
- tiempo desde el último evento

```python
train = obs.loc[obs['Q'] < 20184,]

train['ind'] = (train['USER_ID'].astype(str) + train['Q'].astype(str)).astype(str)
train = train.drop(columns=['USER_ID','Q','next_quarter'])
train = train.set_index('ind')
print(train.shape)
train.head()
```

```python
train.to_csv('train.csv', sep=',', header=True)
```

## Dataset para predecir
Predecimos con los datos de todo 2018 si va a haber una conversión en el primer trimestre del 2019

```python
test = obs.loc[obs['Q'] == 20184,]
test['ind'] = (test['USER_ID'].astype(str) + test['Q'].astype(str)).astype(str)
test = test.drop(columns=['USER_ID','Q','next_quarter'])
test = test.set_index('ind')
print(test.shape)
test.head()
```

```python
test.to_csv('test.csv', sep=',', header=True)
```
