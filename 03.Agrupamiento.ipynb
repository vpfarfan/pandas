# Agrupar por id de usuario, año y mes para el dataset final

##  Probabilidades

```python
# Load data
data = pd.read_csv('./data.csv')
```

```python
grouped_data = data.groupby(['USER_ID','anio','mes','Q']
                            , as_index=False).agg({'PAGE':'count', 'conversion+0':'sum'
                                                   , 'FEC_EVENT':'max', 'gp_prob':'mean'
                                                   , 'dm_prob':'mean','ds_prob':'mean'
                                                   , 'h_prob':'mean', 'new_session':'sum', 'conversion+1':'sum'
                                                   , 'month_q_prob':'mean'})
grouped_data['conversion+0'] = np.where(grouped_data['conversion+0']> 0, 1, 0)
```

```python
grouped_data = grouped_data.groupby(['USER_ID','Q']
                            , as_index=False).agg({'PAGE':'count', 'conversion+0':'sum'
                                                   , 'FEC_EVENT':'max', 'gp_prob':'mean'
                                                   , 'dm_prob':'mean','ds_prob':'mean'
                                                   , 'h_prob':'mean', 'new_session':'sum', 'conversion+1':'sum'
                                                   , 'month_q_prob':'mean'})


grouped_data['conversion+1'] = np.where(grouped_data['conversion+1']> 0, 1, 0)

grouped_data = grouped_data.rename(columns = {'PAGE':'paginas_visitadas', 'conversion+0':'conversiones_previas'
                                              ,'FEC_EVENT':'ultimo_evento','gp_prob':'page_group','dm_prob':'diames'
                                              ,'ds_prob':'diasemana', 'h_prob':'hora'
                                              ,'new_session':'cantidad_sesiones', 'conversion+1':'conversion'})
```

```python
d = {20181: '2018-04-01',
     20182: '2018-07-01',
     20183: '2018-10-01',
     20184: '2019-01-01'
    }
grouped_data['next_quarter'] = grouped_data['Q'].map(d)

# Agregamos el tiempo desde la última transacción hasta el final del cuarto

grouped_data['tiempo_ultimo_evento'] = ((pd.to_datetime(grouped_data['next_quarter'])
                                         - pd.to_datetime(grouped_data['ultimo_evento']))).astype('timedelta64[s]')
print(grouped_data.shape)
grouped_data.head()
```

```python
grouped_data.to_csv('observaciones_Vero.csv',sep=',',header=True)
```

##  Frecuencias

```python
freq = pd.read_csv('./data_Frec.csv')
```

```python
grouped_data = freq.groupby(['USER_ID','anio','mes','Q','session_id'], as_index=False).agg({'prestamo':'sum'
                                                                                            , 'transferencia':'sum'
                                                                           , 'inversiones':'sum', 'solicitud':'sum'
                                                                           , 'quiero':'sum', 'pagos':'sum'
                                                                           , 'eminent':'sum'})

grouped_data = grouped_data.reset_index()
grouped_data.head()
```

```python
grouped_data = grouped_data.groupby(['USER_ID','anio','mes','Q'], as_index=False).agg({'prestamo':'mean'
                                                                                       , 'transferencia':'mean'
                                                                           , 'inversiones':'mean', 'solicitud':'mean'
                                                                           , 'quiero':'mean', 'pagos':'mean'
                                                                           , 'eminent':'mean'})
```

```python
grouped_data = grouped_data.groupby(['USER_ID','Q'], as_index=False).agg({'prestamo':'mean', 'transferencia':'mean'
                                                                           , 'inversiones':'mean', 'solicitud':'mean'
                                                                           , 'quiero':'mean', 'pagos':'mean'
                                                                           , 'eminent':'mean'})
grouped_data.shape
```

```python
grouped_data.to_csv('observaciones_Frec.csv',sep=',',header=True)
```

## Devices

```python
dev = pd.read_csv('./devices_Frec.csv')
dev.head()
```

```python
grouped_data = dev.groupby(['USER_ID','anio','mes','Q','session_id']
#                             , as_index=False
                          ).agg(pages=('CONNECTION_SPEED','count')
                                                   , connection_speed=('CONNECTION_SPEED','mean')
                                                   , is_mobile=('IS_MOBILE_DEVICE','mean')
                                                   , s3=('3','mean'), s2=('2','mean'), s1=('1','mean'), s0=('0','mean'))
grouped_data = grouped_data.reset_index()
grouped_data.head()
```

```python
grouped_data = grouped_data.groupby(['USER_ID','anio','mes','Q']
                            , as_index=False).agg({'pages':'mean', 'connection_speed':'mean', 'is_mobile':'sum'
                                                   , 's3':'sum', 's2':'sum', 's1':'sum', 's0':'sum'})
grouped_data.head()
```

```python
grouped_data = grouped_data.groupby(['USER_ID','Q']
                            , as_index=False).agg({'pages':'mean', 'connection_speed':'mean', 'is_mobile':'sum'
                                                   , 's3':'sum', 's2':'sum', 's1':'sum', 's0':'sum'})
grouped_data.head()
```

```python
grouped_data.to_csv('observaciones_Dev.csv',sep=',',header=True)
```
